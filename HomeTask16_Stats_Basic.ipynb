### Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
q1 = horse_info.pulse.quantile(0.25)
q3 = horse_info.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

outliers = horse_info[(horse_info.pulse > upper_bound) | (horse_info.pulse < lower_bound)].sort_values('pulse')
outliers
```

```python
q1 = horse_info.total_protein.quantile(0.25)
q3 = horse_info.total_protein.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
outliers = horse_info[(horse_info.total_protein < lower_bound) | (horse_info.total_protein > upper_bound)].sort_values('total_protein')
outliers
```

```python
q1 = horse_info.age.quantile(0.25)
q3 = horse_info.age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

outliers = horse_info[(horse_info.age > upper_bound) | (horse_info.age < lower_bound)].sort_values('pulse')
outliers
```

### Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
for empty_values in horse_info.columns:
    quantity = horse_info[empty_values].isnull().sum() 
    percent = quantity / len(horse_info[empty_values])
 
    print(empty_values, ' - ', quantity,  '(', round(percent * 100, 1), '%)', 'пропусков')
```

- surgery : 1 пропуск - исключить
- pulse: 24 пропуска, 8% от общих данных. Заполнение: группировка по возрасту и показателю "Переферический пульс". Заполнение: медианой.
- temperature of extremities: 56 пропусков, 18,7%. Заполним медианным по surgery, age, outcome.
- peripheral pulse: 69 пропусков, что составляет 23,1% от общих данных, что существенно,заполним медианным по pulse.
- pain: так как признак является категориальным, заполним медианным по surgery, age, outcome
- total_protein:11% пропущено. Заполним недостающие данные медианой.
- outcome: 1 пропуск - исключить


```python
df_rezult = horse_info
```

```python
# surgery 
df_rezult = horse_info.dropna(subset=['surgery'])
df_rezult.head()
```

```python
# pulse
df_rezult['pulse'].fillna(horse_info.groupby(['age', 'peripheral pulse'])['pulse'].transform('median'), inplace = True)
df_rezult.head()
```

```python
# temperature of extremities
df_rezult['temperature of extremities'].fillna(horse_info.groupby(['surgery','age','outcome'])['temperature of extremities'].transform('median'), inplace=True)
df_rezult.head()
```

```python
#pain
df_rezult['pain'].fillna(horse_info.groupby(['surgery','age','outcome'])['pain'].transform('median'), inplace=True)
df_rezult.head()
```

```python
# outcome
df_rezult = horse_info.dropna(subset=['outcome'])
df_rezult.head()
```

```python
# peripheral pulse
df_rezult['peripheral pulse'].fillna(horse_info.groupby(['age', 'pulse'])['peripheral pulse'].transform('median'), inplace = True)
df_rezult = horse_info.dropna(subset=['peripheral pulse'])
df_rezult.head()
```

```python
df_rezult['total_protein'].fillna(horse_info['total_protein'].mode()[0], inplace=True)
horse_info.head()
```
